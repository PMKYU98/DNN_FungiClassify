```python
import pandas as pd
```

```python
df = pd.read_csv('data/DATA_benA_4mer_n20.csv')
```

```python
df
```

```python
grouped_df = df.groupby('Species')
downsampled_df = grouped_df.apply(lambda x: x.sample(n=20, random_state=42))  # 각 종별로 한 개의 샘플 선택 (랜덤하게 선택)
downsampled_df.reset_index(drop=True, inplace=True)
downsampled_df
```

```python
downsampled_df.to_csv('DATA_benA_species_4mer.csv')
```
